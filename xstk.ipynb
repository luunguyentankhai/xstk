```python
import os
import shutil
import kagglehub

# Define a function to ignore .nfs files
def ignore_nfs_files(dirname, filenames):
    return [name for name in filenames if name.startswith('.nfs')]

# Pull data
# This downloads the dataset and returns the path to its root directory
path = kagglehub.dataset_download("afumetto/3dprinter")

# Define the path to save raw data
PathFile = "/content/xstk/data/raw"

# Ensure the destination directory is clean before copying
# Remove the directory and its contents if it already exists
if os.path.exists(PathFile):
    shutil.rmtree(PathFile)

# Create the destination directory
os.makedirs(PathFile)

# Copy the dataset from the Kaggle input path to the raw data path
# Use the ignore_nfs_files function to skip problematic .nfs files
shutil.copytree(path, PathFile, ignore=ignore_nfs_files, dirs_exist_ok=True)
```
